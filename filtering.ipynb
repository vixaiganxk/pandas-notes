## Filtering using conditions

```python
#Importing the package and creating the dataframe
import pandas as pd

people = {
    'first_name': ['Magic', 'Old', 'Jack', 'Johnnie', 'Alice', 'Bob', 'Jack', 'Old', 'Eva', 'Frank', 'Alice', 'Jack'],
    'last_name': ['Moments', 'Monk', 'Daniels', 'Walker', 'Smith', 'Monk', 'Lee', 'Miller', 'Clark', 'Davis', 'Clark', 'Andras'],
    'email_ID': ['magicmoments@gmail.com', 'oldmonk@yahoo.com', 'jackdaniel@jack.com', 'johnniewalker@email.com',
                 'alice.smith@example.com', 'bob.johnson@redmail.com', 'jack.lee@example.com', 'old.miller@example.com',
                 'eva.clark@gmail.com', 'frank.davis@expedia.com', 'aliceclar@aws.com', 'jackandras@outlook.com'],
    'contact_no': [9823983242, 4327847238, 6739297423, 9124870439, 8936839800, 6973491820, 9896427123, 7843288324, 5558765432, 8843170957, 7004677862, 8124676874]
}

people_df = pd.DataFrame(people)
people_df
```

### 1. Creating a simple condition

```python
people_df['first_name'] == 'Jack' #---><Return a series of Boolean><---#
```

```python
#Assign the condition to a variable
filt_first_name = (people_df['first_name'] == 'Jack') #--><Series is now assigned to the variable><---#
```

### 1A. Returning the values of the condition from the dataframe

```python
#Passing variable into the dataframe
people_df[filt_first_name] #---><Returns the values of the conditions from the dataframe><---#

#Using Label-location(loc) to return the output
people_df.loc[filt_first_name]
```

### 1B.Returning the specific element using the condition

```python
#Getting a single output using the condition into the series
people_df.loc[filt_first_name, 'email_ID']
```

```python

#Getting the email_ID and contact number based on the condition
people_df.loc[filt_first_name, ['email_ID', 'contact_no']] #---><Returns the output into a dataframe format><---#
```

### 2. AND ['&'], OR ['|'] and NEGATION ['~']

```python
#Creating condition using AND [ & ]
and_con = (people_df['first_name'] == 'Old') & (people_df['last_name'] == 'Monk')

#Creating condition using OR [ | ]
or_con = (people_df['first_name'] == 'Jack') | (people_df['last_name'] == 'Clark')

#Negation---><It's simple just adding [~] in front of the variable creates negation><---#
nand_con = ~and_con
nor_con = ~or_con
```

### 2A. Returning the values based on AND, OR and NEGATION conditions

```python
#Condition:---><first_name is 'Old' and last_name is 'Monk'>---#
people_df[and_con] #---><Returning the output of AND condition><---#
```

```python
#Condition:---><first_name is 'Jack' or last_name is 'Clark'>---#
people_df[or_con] #---><Returning the output of OR condition><---#
```

```python
#Condition:---><first_name is not 'Old' and last_name is not 'Monk'>---#
people_df[nand_con] #---><Returning the output of NAND condition><---#
```

```python
#Condition:---><first_name is not 'Jack' or last_name is not 'Clark'>---#
people_df[nor_con] #---><Returning the output of NOR condition><---#
```

### 3. Reading csv file and performing some conditional operations

```python
file_path = './data/Medals.csv'
medals = pd.read_csv(file_path, encoding='latin1')
medals
```

### 3A. Condition: Filteration based on medals

#### Gold Medals

```python
gold_medal = (medals['Gold'] > 10) 
medals.loc[gold_medal] #---><Returns the dataframe of countires with more than 10 gold medal><----#
medals.loc[gold_medal, ['Team_Country', 'Gold', 'Rank by Total']]
```

#### Silver Medals

```python
silver_medal = (medals['Silver'] > 10) 
medals.loc[silver_medal] #---><Returns the dataframe of countires with more than 10 silver medal><----#
medals.loc[silver_medal, ['Team_Country', 'Silver', 'Rank by Total']]
```

#### Bronze Medals

```python
bronze_medal = (medals['Bronze'] > 10) 
medals.loc[bronze_medal] #---><Returns the dataframe of countires with more than 10 bronze medal><----#
medals.loc[bronze_medal, ['Team_Country', 'Bronze', 'Rank by Total']]
```

#### More than 20 medals as Total

```python
total_medals = medals['Total'] > 20
medals.loc[total_medals] #---><Returns the dataframe of countires with more than 20 total medal><----#
medals.loc[total_medals, ['Team_Country', 'Gold','Silver','Bronze','Total', 'Rank by Total']]
```

#### Less than 10 Medals as Total

```python
total_medals = medals['Total'] < 10
medals.loc[total_medals] #---><Returns the dataframe of countires with more than 50 total medal><----#
medals.loc[total_medals, ['Team_Country', 'Gold','Silver','Bronze','Total', 'Rank by Total']]
```

### 3B. isin() method - To filter out selective countries data

```python
#Assigning the variable of selective countries
countries = ['United States of America', 'India', 'Japan', 'People\'s Republic of China', 'Great Britain', 'Germany']

#Filter condition
coun_filt = medals['Team_Country'].isin(countries)

#Returning the values
medals.loc[coun_filt]
```

### 3C. Using str.contains method to filter a string

```python
str_filt = medals['Team_Country'].str.contains('in') #---><Filters the countries column contains 'in' in them><---#
medals[str_filt]
```
